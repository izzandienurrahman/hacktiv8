```python
import numpy as np
```

```python
# jumlah lemparan koin
n = 50000
```

```python
# head index 0, tail index 1
outcome_head_tail = [0,0]
```

```python
for i in range(n):
    out = np.random.randint(0,2)
    outcome_head_tail[out] = outcome_head_tail[out] + 1
```

```python
outcome_head_tail
```

```python
# Show a pie chart of the results
from matplotlib import pyplot as plt
plt.pie(outcome_head_tail, autopct='%.2f%%', labels=['heads', 'tails'])
plt.legend()
plt.show()
```

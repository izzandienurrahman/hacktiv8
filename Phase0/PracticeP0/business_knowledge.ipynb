```python
import pandas as pd
```

```python
df = pd.read_csv('Mall_customers.csv')
```

```python
df.head()
```

```python
df.groupby('Label').mean()
```

```python
df.groupby(['Gender','Label']).mean()
```

```python
df_income_to_spending_ratio = df['Annual Income (k$)']/df['Spending Score (1-100)']
```

```python
df['inout ratio'] = df_income_to_spending_ratio #larger = more thrifty/frugal. lower = lavish/consumptive
```

```python
df.groupby(['Gender','Label']).mean().sort_values(by='inout ratio') #dengan beda gender
```

```python
df.groupby(['Label']).mean().sort_values(by='inout ratio')
```
